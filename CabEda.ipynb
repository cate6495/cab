Merging the Datasets

```python
customertransactiondataset=pd.merge(customerdataset, transactiondataset,on='Customer ID')
customertransactioncabdata=pd.merge(customertransactiondataset, cabdataset, on='Transaction ID')
data=pd.merge(customertransactioncabdata, citydataset, on='City')
```

```python
print("Dataset length :: ",len(data))
print("Dataset shape :: ", data.shape)
data.head(5)
```

```python
data.rename(columns={'Transaction ID': 'Transactionid', 'Customer ID': 'Customerid'}, inplace=True)
data.rename(columns={'Income (USD/Month)': 'Income', 'Price Charged': 'Price'}, inplace=True)
data.rename(columns={'KM Travelled': 'Kmtravelled', 'Cost of Trip': 'Cost'}, inplace=True)
```

```python
data.info()
```

```python
# converting the date of travel from object to datetime
data['Date of Travel']=pd.to_datetime(data['Date of Travel'])
```

```python
#separating date of travel to month and year
months=[]
years=[]
for i in range(len(data['Date of Travel'])):
    months.append(data['Date of Travel'][i].month)
    years.append(data['Date of Travel'][i].year)
data['Month']=months
data['Year']=years
```

```python
#dropping the column date of travel
data.drop('Date of Travel',axis=1,inplace=True)
```

HYPOTHESIS TESTING

There is a difference in margins depending on the mode of payment

```python
#Yellow Cabs
a = data[(data.Payment_Mode=='Cash')&(data.Company=='Yellow Cab')].groupby('Transactionid').Margin.mean()
b = data[(data.Payment_Mode=='Card')&(data.Company=='Yellow Cab')].groupby('Transactionid').Margin.mean()
_, p_value = stats.ttest_ind(a.values,b=b.values,equal_var=True)
if(p_value<0.05):
    print('We accept alternate hypothesis that theres a difference')
else:
    print('We accept null hypothesis that theres no difference')
    
print('P value is ', p_value)
```

There is no difference in margin for Yellow cab regardless of the mode of payment

```python
#Pink Cabs
a = data[(data.Payment_Mode=='Cash')&(data.Company=='Pink Cab')].groupby('Transactionid').Margin.mean()
b = data[(data.Payment_Mode=='Card')&(data.Company=='Pink Cab')].groupby('Transactionid').Margin.mean()


_, p_value = stats.ttest_ind(a.values,b=b.values,equal_var=True)
if(p_value<0.05):
    print('We accept alternate hypothesis that theres a difference')
else:
    print('We accept null hypothesis that theres no difference')
    
print('P value is ', p_value)
```

There is no difference in margin for Pink Cab regardless of the mode of payment
